## Starting with Spotify API

```python
playlist = sp.user_playlist_tracks("spotify", playlist_id = "37i9dQZEVXbOa2lmxNORXQ", market="GB")
```

```python
#pprint.pprint(playlist)
```

```python
tracks = playlist['items']
len(tracks) # number of tracks in the playlist
```

```python
# List to store track names and ids
track_list = []

# Loop through the items and extract track names and ids
for track_item in tracks:
    track_name = track_item['track']['name']
    track_id = track_item['track']['id']
    track_list.append({'track_name': track_name, 'track_id': track_id})

#pprint.pprint(track_list)
```

```python
# List of track IDs and names from the previous step
track_ids = [track['track_id'] for track in track_list]
track_names = [track['track_name'] for track in track_list]
```

```python
# Spotipy method to get audio features for each track
audio_features = sp.audio_features(track_ids)

# pprint.pprint(audio_features)
```

```python
# List to store the relevant features for each track
track_features_list = []

# Loop through the audio features response and add the track names
for i, feature in enumerate(audio_features):
    # Create a dictionary for track features and include the track name
    track_features = {**feature, 'name': track_names[i]}  # Merge all feature keys and add track name
    # Add the track features to the list
    track_features_list.append(track_features)
```

```python
# Convert the track features list to a DataFrame
df_tracks = pd.DataFrame(track_features_list)
```

```python
# save the DF to a csv file
df_tracks.to_csv("data/clean/tracks_0.csv", index=False)
print("Saved tracks_0.csv")
```

 ## Get song features for multiple playlist

```python
playlist_ids = [
    "37i9dQZEVXbOa2lmxNORXQ",
    "37i9dQZF1DXcBWIGoYBM5M",  
    "37i9dQZF1DX0XUsuxWHRQd",  
    "37i9dQZEVXbJiZcmkrIHGU",  
    "487jKTFqWhs6b0AEUz0WpX",  
    "061xZeYgr9dCCYh2MWovDs",  
    "37i9dQZF1DWXRqgorJj26U",  
]

# List to store all track features
all_track_features = []

# Loop through each playlist ID
for index, playlist_id in enumerate(playlist_ids):
    # Fetch tracks from the playlist
    playlist = sp.user_playlist_tracks("spotify", playlist_id, market="GB")
    tracks = playlist['items']

    # Extract track names and IDs
    track_list = []
    for track_item in tracks:
        track_name = track_item['track']['name']
        track_id = track_item['track']['id']
        track_list.append({'track_name': track_name, 'track_id': track_id})

    # List of track IDs from the previous step
    track_ids = [track['track_id'] for track in track_list]
    track_names = [track['track_name'] for track in track_list]

    # Spotipy method to get audio features for each track
    audio_features = sp.audio_features(track_ids)

    # Loop through the audio features response and add the track names
    for i, feature in enumerate(audio_features):
        if feature is not None:  # Check if feature data is available
            # Create a dictionary for track features, including all features and the track name
            track_features = {**feature, 'track_name': track_names[i]}
            all_track_features.append(track_features)

# Convert the accumulated track features list to a DataFrame
df_tracks = pd.DataFrame(all_track_features)

df_tracks.head()
```

```python
# Save the df to a csv file
df_tracks.to_csv("data/clean/all_track_features.csv", index=False)
```
