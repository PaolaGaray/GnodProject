## Clustering Models

```python
# X = StandardScaler().fit_transform(X)
```

### K-Means

```python
X = spotify_df[["danceability","energy","loudness","speechiness","acousticness", "instrumentalness","liveness","valence","tempo","duration_ms"]]
```

**2 Clusters**

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
pred = kmeans.predict(X)
```

```python
silhouette_score(X, pred, metric='euclidean')
```

```python
pred
```

```python
kmeans.cluster_centers_
```

#### Yellowbrick

```python
model = cluster.KMeans()
visualizer = KElbowVisualizer(model, k=(2,10))
visualizer.fit(X)
visualizer.poof()
```

**4 Clusters**

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
pred = kmeans.predict(X)
```

```python
silhouette_score(X, pred, metric='euclidean')
```

```python
kmeans.cluster_centers_
```

**3 Clusters**

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
pred = kmeans.predict(X)
silhouette_score(X, pred, metric='euclidean')
```

```python
kmeans.cluster_centers_
```

## Agglomerative clustering

**Linkage Ward**

```python
agglomerative = cluster.AgglomerativeClustering(n_clusters=3,linkage='ward')
pred = agglomerative.fit_predict(X)
```

```python
print("Model 2 Silhouette Score: {}".format(silhouette_score(X, pred, metric='euclidean')))
```

**Linkage Single**

**3 Clusters**

```python
agglomerative = cluster.AgglomerativeClustering(n_clusters=3,linkage='single')
pred = agglomerative.fit_predict(X)
```

```python
print("Model 2.single Silhouette Score: {}".format(silhouette_score(X, pred, metric='euclidean')))
```

```python
pred
```

**4 Clusters**

```python
agglomerative = cluster.AgglomerativeClustering(n_clusters=4,linkage='single')
pred = agglomerative.fit_predict(X)
```

```python
print("Model 2.single Silhouette Score: {}".format(silhouette_score(X, pred, metric='euclidean')))
```

## Using KMeans clustering w/ 4 Clusters

```python
non_stand_spot_df = spotify_df
```

```python
non_stand_spot_df['cluster'] = pred
```

```python
non_stand_spot_df.head()
```

```python
non_stand_spot_df['cluster'].value_counts()
```

```python
# exporting the non-standardized df
non_stand_spot_df.to_csv('non_stand_spot_df.csv', index=False)
```
